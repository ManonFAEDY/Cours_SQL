## Introduction
**SQL** (=**_Structured Query Language_**) : Language informatique permettant d'exploiter des bases de données relationnelles en faisant des requêtes

**Base de Donnée Relationnelle** ("_Relational Database_" ou _RDB_) = Système dans lequel l'**information** (les données) est organisée sous forme de **tables** (tableaux à 2 dimensions) qui sont liées entre elles grâce à une identifiant unique appelé **Clé Primaire** ("_Primary Key_") dans la table principale et Clé Etrangère ("_Foreign Key_") dans les autres tables.
> Avantages : données restent cohérentes, structurées et faciles à trouver

### Les types de données

- INT ou INTEGER : Nombres entiers
- FLOAT : Nombres à virgule
- VARCHAR (CHARACTER VARYING) : Texte variable (jusqu'à n caractères)
- STRING : Très long texte
- DATE / DATETIME : Dates et heures

## A. Création de table et sélection de donnéee

- **CREATE TABLE** : créer une table

```sql
CREATE TABLE nom_de_la_table (
    colonne_id type_donnees,
    variable_1 type_donnees,
    variable_2 type_donnees);
```

- **ALTER TABLE ... ADD COLUMN** : ajouter une colonne à une table, par défaut toutes les lignes existantes auront la valeur NULL dans cette nouvelle colonne.

`ALTER TABLE nom_de_la_table ADD COLUMN nom_de_la_nouvelle_colonne type_donnees;`

- **INSERT INTO ... VALUES** : ajouter de nouvelles lignes

`INSERT INTO nom_de_la_table VALUES (val1, val2...);`

- **UPDATE ... SET ... WHERE** : modifier des données existantes

`UPDATE nom_de_la_table SET nom_de_la_colonne = nouvelle_valeur WHERE condition;`

Attention sans le WHERE cela va changer les valeurs de TOUTE la colonne.

- **DELETE FROM ... WHERE** : supprimer des lignes

`DELETE FROM nom_table WHERE condition;`

- **SELECT ... FROM** : Selectionner des colonnes à partir d'une table 

`SELECT nom_de_la_colonne1, nom_de_la_colonne2 FROM nom_de_la_table;`

Séléctionner toutes les colonnes d'une table avec *

`SELECT * FROM nom_de_la_table`

- **SELECT ... AS** : Renommer ou créer des colonnes / variables à partir d'autres colonnes

`SELECT nom_de_la_colonne1 AS nouveau_nom_de_la_colonne1 FROM nom_de_la_table;`

`SELECT (nom_de_la_colonne1 + nom_de_la_colonne2) AS nom_de_la_nouvelle_colonne FROM nom_de_la_table;`

- La méthode **COPY ... FROM** : méthode la plus rapide pour charger des millions de lignes d'un coup. Vous devez d'abord avoir créé la structure de la table vide. Méthode valide pour PostgreSQL mais pas Databricks.

```sql
COPY ma_table 
FROM '/chemin/vers/votre/fichier.csv' 
DELIMITER ',' 
CSV HEADER;
```

---
- La méthode **CREATE TABLE ... AS SELECT** (CTAS) : sans pré-création de la table. Databricks va lire le fichier et deviner quels sont les types de données.

```sql
CREATE TABLE video_games_sales
AS SELECT * FROM read_files(
  '/Workspace/Users/manon.faedy@gmail.com/vgsales.csv',
  format => 'csv',
  header => true,
  inferSchema => true  -- <--- C'est l'option magique
);
```

L'option inferSchema => true analyse le fichier et applique les types automatiquement. Il peut y avoir des erreurs donc nécessaire de vérifier après.

---
- La méthode **COPY INTO ... FROM** : plus robuste et gère mieux les erreurs dans Databricks. Nécessite de pré-créer la table.

```sql
COPY INTO video_games_sales
FROM '/Workspace/Users/manon.faedy@gmail.com/vgsales.csv'
FILEFORMAT = CSV
FORMAT_OPTIONS ('header' = 'true', 'inferSchema' = 'true')
```

## C. Tri et filtrage de données

A placer après le `SELECT * FROM Nom_de_la_table` ...

- `WHERE` : doit remplir la condition
    - exemple : `WHERE Pays = "France"`
    - exemple : `WHERE Pays != "USA"`

  - `OR` et `AND` : filtrer en utilisant des conditions. 🚨 Attention : Opérateurs logiques prioritaires donc utilisation des parenthèses dans les WHERE.

      - Exemple : `WHERE (prix < 20 OR promo = 1) AND stock > 0`

  - `LIKE ` : filtrer en utilisant un pattern avec un ou plusieurs caractères
    - `LIKE "%text%"` : peu importe ce qu'il y a devant ou derrière
    - `LIKE "%text"` : peu importe ce qu'il y a devant
    - `LIKE "text%"` : peu importe ce qu'il y a derrière
    - `LIKE "PS_"` : si on cherche des lettres à des endroits précis (chaque tiret est une lettre)

  - `IN` et `BETWEEN` : filtrer à partir d'une liste ou intervalle

    - `IN ('PS5', 'Xbox', 'PC')` : Plus propre que des OR successifs.

    - `BETWEEN 20 AND 50` : Pour les prix et les dates.

- `ORDER BY` : ASC pour un tri alphabétique (A-Z) et chiffre croissant, ou DESC pour l'inverse
  - exemple : `ORDER BY prix DESC`

- `LIMIT` : seulement un certain nombre de ligne
  - exemple : `LIMIT 3` pour afficher les 3 premières lignes

- Le concept de N/A : N/A ou NULL n'est pas 0 ni un espace vide, c'est une "absence de valeur".

    Syntaxe : `WHERE colonne IS NULL` ou `IS NOT NULL`
