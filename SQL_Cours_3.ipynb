## Introduction : Architecture Relationnelle et Jointures

- Clé Primaire (PK) : L'identifiant unique d'une table (ex: id_client).

- Clé Étrangère (FK) : La référence d'une table vers une autre (ex: id_client présent dans la table Commandes).

- Le concept de Normalisation : Pourquoi on ne stocke pas le nom du client dans la table commande (éviter la redondance).

![image_1772820747514.png](./image_1772820747514.png "image_1772820747514.png")


- INNER JOIN : Le point commun. On ne garde que ce qui matche des deux côtés.

- LEFT JOIN : On garde tout ce qui est à gauche (ex: tous les clients) + les infos de droite si elles existent (leurs commandes). Si pas de commande, on aura du NULL.

## A. INNER JOIN

On conserve uniquement les lignes qui sont communes au 2 tables.

```sql
SELECT *
FROM table_A
INNER JOIN table_B ON table_A.colonne_Foreign_Key = table_B.colonne_Primary_Key
```

Si on veut séléctionner uniquement certaines colonnes pour chaque table il faut le préciser en spécifier de quelle table est issue chaque colonne :

```sql
SELECT table_A.colonne_1, table_A.colonne_2, table_B.colonne_1
FROM table_A 
INNER JOIN table_B ON table_A.colonne_Foreign_Key = table_B.colonne_Primary_Key
```

ou avec l'utilisation d'alias pour rendre ça plus lisible :

```sql
SELECT a.colonne_1, a.colonne_2, b.colonne_1
FROM table_A AS a
INNER JOIN table_B AS b ON a.colonne_FK = b.colonne_PK
```

## B. LEFT JOIN

On conserve toutes les lignes de la première table et on leur rajoute des informations à partir de la deuxième table (une absence d'information donnera une valeur NULL).

```sql
SELECT *
FROM table_A
LEFT JOIN table_B ON table_A.colonne_Foreign_Key = table_B.colonne_Primary_Key
```

Plusieurs colonnes avec alias :
```sql
SELECT a.colonne_1, a.colonne_2, b.colonne_1
FROM table_A AS a
LEFT JOIN table_B AS b ON a.colonne_FK = b.colonne_PK
```

## C. Les conditions : CASE...WHEN


Création d'une nouvelle colonne à partir de conditions dans d'autres colonnes :
```sql
CASE 
    WHEN condition_1 THEN 'Résultat A'
    WHEN condition_2 THEN 'Résultat B'
    ELSE 'Résultat par défaut'
END AS nom_de_ma_nouvelle_colonne
```
